# Pandas plotting

```python
# importing the required(important) libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# lets create a data 
a = np.random.random((4,4))
a
```

```python
d = pd.DataFrame(a,columns=['a','b','c','d'])
d
```

## Line Plot :

```python
# by default its a line plot 

#d.plot()
d.plot(kind = 'line') # same output as above
plt.xlabel('X-axis data')
plt.ylabel('Y-axis data')
plt.title('line plot using pandas')
```

```python
# now, separate each plot 
d.plot(kind='line',subplots=True)
```

```python
# now change the layout of the graph 
d.plot(kind='line', subplots=True, layout=(2,2))
```

```python
#increase the figure size
# change layout
d.plot(kind='line',subplots=True,layout=(2,2),figsize=(8,6))
plt.show()
```

```python
# line plot: c vs d
# column vs column
d.plot(x='c',y='d')
```

```python
d
```

## Bar plot

```python
# Transpose (T)
d.T.plot(kind='bar')
```

```python
d.plot(kind='bar')
```

```python
# On X axis above we have 0,1,2,3 
# i want to change labels from number to word Ex. col_1 col_2...
d.plot(kind='bar')
#plt.xticks((0,1,2,3),('col_1','col_2','col_3','col_4'))
#change rotation
plt.xticks((0,1,2,3),('col_1','col_2','col_3','col_4'),rotation=45)
plt.show()
```

```python
# stacked bar plot
d.plot(kind='bar',stacked=True)
#plt.xticks((0,1,2,3),('col_1','col_2','col_3','col_4'))
#change rotation
plt.xticks((0,1,2,3),('col_1','col_2','col_3','col_4'),rotation=45)
plt.show()
```

```python
d.plot(kind='bar',subplots=True,layout=(2,2))
#plt.xticks((0,1,2,3),('col_1','col_2','col_3','col_4'))
#change rotation
plt.xticks((0,1,2,3),('col_1','col_2','col_3','col_4'),rotation=45)
plt.show()
```

```python
# 2 plotting versions are available
#d.plot(kind='bar') its one option
d.plot.bar() #its another approach
```

## Pie plot :

```python
d.plot(kind='pie',subplots=True,figsize=(12,10))
plt.show()
```

```python
d.plot.pie(subplots=True)
```

```python
# add % data in each pie
d.plot(kind='pie',autopct='%.f%%',subplots=True,figsize=(12,10),layout=(2,2),
      explode=[0.2,0,0,0])
plt.show()
```
